```python
import time
import pymongo
from pymongo import MongoClient
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

client = MongoClient('192.168.142.99', 27017)
dbname = ['news','blog','forum','social_media']
collectionname = ['Applenews_post','Ettoday_post','Setn_post','Pixnet_post','Ck101_post','Dcard_post','Eyny_post',\
                  'Mobile01_post','Ptt_post','FB_post']

texts = []
for dbn in dbname:
    for coln in collectionname:
    
        db = client[dbn]
        collection = db[coln]

        cursor = collection.find()
        cursor.count()
        ## 進入每篇 post 進行分析
        for document in cursor:
            texts.append(document["newcontent_tag"].split('|'))

# del texts[663]
print (texts[:10])
```

```python
from gensim import corpora, models, similarities

#### step-1 : 將 list中的詞項轉換为 word to id 映射字典 bag of words词袋
dictionary = corpora.Dictionary(texts)
# print (dictionary)
# print (dictionary.token2id)
```

```python
#### step-2 : 將文本轉成詞項對應編號的語料庫
# dictionary核心函数 doc2bow
# 將文本轉化成依照字典中词的编号和频数
# tuple style:(token_id,token_count)

corpus = [dictionary.doc2bow(text) for text in texts]
# print (corpus[:10])
```

```python
#### step-3 : 將語料庫轉成 詞項-文檔矩阵(TDM)

# 語料庫和 numpy可以相互兼容，在轉換時需要设置特征函数
# matutils math utils 
# gensim.matutils.corpus2dense(corpus, num_terms, num_docs=None, dtype=<type 'numpy.float32'>) 
# 将複雜的 corpus 轉換成稀疏 corpus (流式) --> TDM 矩陣
# 流式数据的优点：占用内存小。

from gensim import matutils

print (len(dictionary))
numpy_matrix = matutils.corpus2dense(corpus, num_terms=len(dictionary) )    # num_terms = tokens 數量
print (numpy_matrix[:10,])

# corpus = gensim.matutils.Dense2Corpus(numpy_matrix) --> 反轉的概念
# 詞項-文檔矩阵(TDM)，其中值为词频
```

```python
#### step-4 : TFIDF 處理 ==> corpus_tfidf
# class gensim.models.tfidfmodel.TfidfModel(corpus=None, id2word=None, dictionary=None, wlocal=<function identity>, 
#                                            wglobal=<function df2idf>, normalize=True)
# TFIDF 計算函數
tfidf = models.TfidfModel(corpus)
print (tfidf)

# 轉成 TFIDF 矩陣
corpus_tfidf = tfidf[corpus]

# 观察结果
# for doc in corpus_tfidf[:10]:
#     print (doc)
```

```python
#### step-5 : run LDA (Latent Dirichlet Allocation) 主題模型 

# 設定主提挖掘數量
num_topics = 10

import time
# 記錄開始時間 Record start time : 約 15秒
tStart = time.time()

lda = models.LdaModel(corpus_tfidf,id2word=dictionary,num_topics=num_topics )
# lda_alpha = models.LdaModel(corpus_tfidf, id2word=dictionary, num_topics=num_topics, alpha='auto', eval_every=5)  
# train asymmetric alpha from data

print ('done')

# 記錄結束時間 Record stop time
tStop = time.time()
print(tStop - tStart)
```

```python
#### 運用 2 : 在主题空间比较两两文档之间的相似度
import numpy as np
dense = np.zeros((len(corpus_lda),30),float)    # 10 --> num_dopics 主題數 , dense = 1019*10 array

# 每一次的 corpus_lda 在小數位後都有些許不同 --> 導致最相似文章有些不同
for ti,t in enumerate(corpus_lda):
    for tj,v in t:
        dense[ti,tj] = v
              
#接下来计算文档相似度--得到矩阵中文档间距离的方阵
from scipy.spatial import distance

# distance.squareform:用于将 v向量(长度为d*(d-1)/2）转化为对角线全为零的方阵 --> 1019*1019 array
# distance.pdist:pdist(X):计算两两行向量之间的欧几里得距离
pairwise = distance.squareform(distance.pdist(dense))
# print (len(pairwise[0]))

#将对角线元素全部置为比矩阵中最大值还要大， 目的在計算最近距離時不要取到本身文本
largest = pairwise.max()
# print (largest)
for ti in range(len(corpus_lda)): 
     pairwise[ti,ti] = largest + 1

# count = 0
# for w in pairwise[5]:
#     print (w)
#     count +=1
# print (len(pairwise))    

## 自訂函數找到與搜尋文章 在主題概念上最接近的文章 
def closest_to(doc_id):
     return pairwise[doc_id].argmin()

print (closest_to(232))
#darray.argmin(axis=None, out=None)
#Return indices of the minimum values along the given axis of a.
#Refer to numpy.argmin for detailed documentation.
```
